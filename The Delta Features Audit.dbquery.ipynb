```sql
-- VALIDATION 3: DELTA HISTORY AUDIT
-- This query retrieves the recent history of operations performed on the gold_fact_sales Delta table.
-- Look for 'OPTIMIZE', 'VACUUM', and 'SET ROW FILTER' in the operation list to audit maintenance and security actions.

SELECT 
    version,                  -- Table version number
    timestamp,                -- Time when the operation was performed
    operation,                -- Type of operation (e.g., OPTIMIZE, VACUUM, SET ROW FILTER)
    operationParameters       -- Additional details about the operation
FROM (DESCRIBE HISTORY ecommerce_analytics_dev.gold_layer.gold_fact_sales)
LIMIT 20;                     -- Show the 20 most recent operations
```
